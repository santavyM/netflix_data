```python
import pandas as pd

# Načti CSV (změň cestu pokud máš soubor jinde)
df = pd.read_csv("netflix_titles.csv")

# Zobraz prvních 5 řádků
df.head()
```

```python
# Počet přidaných titulů podle roku
df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,6))
```

```python
import matplotlib.pyplot as plt

df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,6))
plt.title("Počet titulů podle roku vydání")
plt.xlabel("Rok")
plt.ylabel("Počet titulů")
plt.grid(True)
plt.show()
```

```python
# Vytvoříme nový sloupec, kde každý žánr bude zvlášť
genre_series = df['listed_in'].dropna().str.split(', ').explode()

# Spočítáme nejčastější žánry
genre_series.value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 nejčastějších žánrů")
plt.xlabel("Žánr")
plt.ylabel("Počet titulů")
plt.grid(True)
plt.show()
```
